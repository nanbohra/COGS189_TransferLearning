```python
# Simple MLP implementation + classifier head

class MLP(nn.Module):
    def __init__(self, n_input, n_hidden1=64, n_hidden2=32, dropout=0.2):
        super().__init__()

        self.feature_extractor = nn.Sequential(
            nn.Linear(n_input, n_hidden1)
            nn.ReLU(),
            nn.Linear(n_hidden1, n_hidden2),
            nn.Dropout(p=dropout),
            nn.ReLU()
        )

        self.classifier = nn.Linear(n_hidden2, 2) # this will be fine-tuned

    def forward(self, x):
        # x is csp features
        features = self.feature_extractor(x)
        logits = self.classifier(features)
        probs = F.softmax(logits, dim=1)
        return probs
```
